```python
def change_party(prefid):  
    copy_list = []
    pref = df.loc[df["prefid"] == prefid]

    distnums = pref["distnum"][-1:]
    distnum = 1

    while distnum <= int(distnums):
        dist = pref[pref['distnum'].isin([int(distnum)])]
        #選挙区別に分けた
        
        if any(dist["party"] == 1) == 1 or any(dist["party"] == 3) == 1:
            gp1 = dist[(dist["party"] == 1) | (dist["party"] == 3)]
            if len(gp1) > 1:
                pass
            else:
                dist.loc[dist["party"] == 1, "party"] = "X"
                dist.loc[dist["party"] == 3,"party"] = "X"
    
        if any(dist["party"] == 2) == 1 or any(dist["party"] == 4) or any(dist["party"] == 5) == 1:
            oplist1=[2,4,5]
            opposition_party1 = dist[(dist["party"] == 2) | (dist["party"] == 4) | (dist["party"] == 5)]
            #print(opposition_party1)
        
            if len(opposition_party1) >= 1:
                for op1 in oplist1:
                    dist.loc[dist["party"] == op1, "party"] = "Y"
        
            else:
                pass
        
        if any(dist["party"] == 35) == 1:
            dist.loc[dist["party"] == 35, "party"] = "Z"
        
        dist_copy = dist.copy()
        copy_list.append(dist_copy)
        distnum += 1
        
    changed = pd.DataFrame()
    for ward in copy_list:
        changed = changed.append(ward)
        
    return changed
```

## 同選挙区内の同じ勢力の票をすべて合計してまとめる
> win,votes,vshareは書き換えが必要

とりあえず自公連立はどの選挙区でも1人しかいない(維新も)
> 民主党+共産党+社民党を足すだけ

1. Yが複数の立候補者を同一選挙区に立てている場合
> 1. 空の1×9のdataframeにカラム名を割り振り、partyのところにYの獲得した票数の合計を代入
> 1. Y勢力の行をpref県のdataframeから削除する
> 1. 県のdataframeの残った部分を成形し、先ほど作ったdataframeにappend

1. Yが1人
> 1. 県のdataframeを成形

```python
def parallel_pref(prefid):
    parallel_list = []
    
    copy = change_party(prefid).copy()
    distnums = copy["distnum"][-1:]
    distnum = 1

    while distnum <= int(distnums):
        dist = copy[copy['distnum'].isin([int(distnum)])]
        
        parallel_y = np.random.rand(1,9)
        parallel_y = pd.DataFrame(parallel_y)
        parallel_y.columns = ["prefid","pref","distnum","party","win","votes","tvotes","vshare","rank"]
        parallel_y = parallel_y[parallel_y[:] ==0]
        #len(copy[copy['distnum'].isin([int(distnum)])])-1がその選挙区における野党勢力の数
        if len(dist[dist['party'].isin(["Y"])]) >1:
            #勢力Yをひとまとめにする
            yatou_y = dist[dist['party'].isin(["Y"])]
            parallel_y["votes"] = sum(yatou_y["votes"])
            parallel_y.loc[:,["prefid","pref","distnum","party","tvotes"]] = [prefid,copy.iat[0,3],distnum,"Y",dist["tvotes"][-1:]]
            
            #勢力Yを削除する
            noy_pref = dist[dist['party'] != "Y"].copy()
            noy_pref.drop(['id', 'name',"dual","inc","age"], axis=1,inplace = True)
            noy_pref["rank"] = 0
            parallel = noy_pref.append(parallel_y)
        else:
            #distの余分な列を削除する
            parallel = dist.copy()
            parallel.drop(['id', 'name',"dual","inc","age"], axis=1,inplace = True)
            parallel["rank"] = 0
            
        #parallelのwin,vshare,rankを処理
        parallel.loc[:,["win","vshare","rank"]] = [0,0,0]
        parallel["vshare"] = parallel["vshare"].astype(float)
        parallel = parallel.sort_values("votes",ascending=False)
        parallel.iat[0,4] = 2
        parallel.loc[:,"rank"] = parallel.loc[:,"votes"].copy().rank(ascending=False)
        for i in range(len(parallel)):
            #votes/tvotes
            parallel.iat[i,7] = (int(parallel.iat[i,5]) / int(parallel.iat[i,6])) *100
        
        parallel_copy = parallel.copy()
        parallel_list.append(parallel_copy)
        distnum += 1
        
    isekai = pd.DataFrame()
    for ken in parallel_list:
        isekai = isekai.append(ken)
    isekai = isekai.reset_index(drop=True)
    return isekai
```

## 異世界のefgpを求める

```python
def japan_efgp2014_parallel(prefid):    
    pref = parallel_pref(prefid)
    distnums = pref["distnum"][-1:]
    
    #空のデータフレームを各都道府県のcokumn数＝立候補者数,row数=政党の数で作成
    void = np.random.rand(int(distnums),17)
    void = pd.DataFrame(void)
    void.columns = ["1","2","3","4","5","33","34","35","98","99","100","101","102","103","X","Y","Z"]
    void = void[void[:] ==0]

    distnum = 1

    while distnum <= int(distnums):
        #dist = pref[pref["distnum"] == distnum]
        dist = pref[pref['distnum'].isin([int(distnum)])]
        #選挙区別に分けた
    
        for party in [1,2,3,4,5,33,34,35,98,99,"X","Y","Z"]:
            #pref28兵庫でエラーinvalid type comparison<-solved
            dist_each_party = dist[dist['party'].isin([party])]
            #dist_each_party = dist[dist["party"] == party]
            #<-各選挙区の候補者数
        
            #row=0でpass
            if len(dist_each_party) == 0:
                pass
            elif len(dist_each_party) >1:
                #ある勢力が1選挙区に複数人候補者を立てた場合<-98,99の場合
                mul_ca=1
                #98,99がまとめて1つの勢力とみなされるのを防ぐため、便宜的にpartyの番号を+1している
                #エラー福岡40でその他勢力が4つもいた<-政党番号102に対応させたsolved
                while mul_ca <= len(dist_each_party):
                    ex99 = dist_each_party[mul_ca-1:mul_ca]
                    if all(ex99["win"] == 2) == 1:
    
                        wasted = dist["votes"].iat[0] - dist["votes"].iat[1]
                        void[str(mul_ca-1 + party)][distnum-1] = wasted
    
                    elif all(ex99["win"] == 2) ==0:
                        wasted=ex99["votes"]
                        void[str(mul_ca + party)][distnum-1]=wasted
                    mul_ca +=1
            else:
                if all(dist_each_party["win"] == 2) == 1:
                    wasted = dist["votes"].iat[0] - dist["votes"].iat[1]

                    void[str(party)][distnum-1] = wasted
            
                elif all(dist_each_party["win"] != 2) == 1:
                    wasted=dist_each_party["votes"]

                    void[str(party)][distnum-1] =wasted
                
                else:
                    pass
                
        distnum += 1
    void
    #voidに各勢力の死票数が入力された
    totalvotes = pref[pref["win"] == 2]
    totalvotes = sum(totalvotes["tvotes"])
    void = void.dropna(how='all', axis=1)
    void = void.fillna(0)
    #各党の総死票数の計算
    nwv = void.sum()
    dict_party = {"1":"自民党","2":"民主党","3":"公明党","4":"共産党","5":"社民党","33":"生活の党","34":"次世代の党",
                  "35":"日本維新の会","98":"その他・無所属","99":"その他・無所属","100":"その他・無所属",
                  "101":"その他・無所属","102":"その他・無所属","X":"自公連立","Y":"野党共闘(仮想)","Z":"維新"}
    #dataframe"kong"にefficiency gapの値を入力する
    kong = np.random.rand(len(void.columns),len(void.columns))
    kong = pd.DataFrame(kong)
    kong.columns = void.columns
    kong.index = void.columns
    kong = kong[kong[:] ==0]
    #efficiency gapの計算
    j = 1
    k = 1
    while j < len(nwv):   
        efgp_ep = []
        while k <= len(nwv): 
            #nwvのリストの一つ後から一つ前を引く<=野党系が一つ後になる
            dif = nwv[j-1] - nwv[k]
            efgp_ep.append(dif)
            kong.iat[j-1,k] = dif/totalvotes*100
        
            k +=1    
        
            if k==len(nwv):
                break
        efgp = list(map(lambda x: x / totalvotes *100, efgp_ep))

        k = j
        j += 1
        
    kong = kong.rename(columns = dict_party,index = dict_party)
    return kong
```
